```python
colorida = {'a':3, 'b': 6, 'c':3, 'd':8, 'e':9,'f':5,'g':10,'h':9,'i':8,'j':4,'k':4,'l':9,'m':8,'n':8, 'o':10 ,'p':4,'q':7,'r':6,'s':5,'t':4,'u':8,'v':5,'w':4,'x':7,'y':5,'z':11}
dourada = {'a':17, 'b': 16, 'c':13, 'd':16, 'e':20,'f':10,'g':17,'h':15,'i':26,'j':24,'k':8,'l':20,'m':16,'n':25, 'o': 17, 'p':22,'q':24,'r':16,'s':24,'t':16,'u':22,'v':16,'w':16,'x':27,'y':16,'z':21}
numeros = {'0':10,'1':10,'2':10,'3':10,'4':10,'5':10,'6':10,'7':'0','8':10,'9':10}
```

Contei todas as miçangas e essa foi a quantidade obtida de cada letra. Mas no dia sugiro que alguém reconte só pra reforçar.

## Colorida

```python
# consigo fazer a pulseira?

palavra = input('digite a palavra/frase que quer confeccionar: ')
palavra = palavra.replace(' ','')
p = {}
n = {}
for letra in palavra:
  if letra.isdigit() and letra not in n:
    n[letra] = palavra.count(letra)
  elif letra not in p:
    p[letra] = palavra.count(letra)

consc = 1
consn = 1
for letra in p:
  if p[letra] > colorida[letra]:
    print('não consigo')
    consc = 0
    break

for num in n:
  if n[num] > numeros[num]:
      print('não consigo')
      consn = 0
      break

if consc == 1 and consn == 1:
  print('consigo')
```

```python
# fazer a pulseira e diminuir quantidade de letras disponíveis

if consc == 1 and consn == 1:
  print(palavra)
  conf = input('fazer pulseira com essa palavra? digite s para sim ou qualquer coisa para nao ')
  if conf == 's':
    print('quantidades atualizadas: ')
    for letra in p:
      colorida[letra]-+p[letra]
      print(letra + ': ' + str(numeros[num]))
    for num in n:
      numeros[num]-+n[num]
      print(num + ': ' + str(numeros[num]))
  else:
    print('ok')
```

## Dourada

```python
# consigo fazer a pulseira?

palavrad = input('digite a palavra/frase que quer confeccionar: ')
palavrad = palavrad.replace(' ','')
dp = {}
nd = {}
for letra in palavrad:
  if letra.isdigit() and letra not in nd:
    nd[letra] = palavrad.count(letra)
  elif letra not in dp:
    dp[letra] = palavrad.count(letra)

consd = 1
consnd = 1
for letra in dp:
  if dp[letra] > dourada[letra]:
    print('não consigo')
    consd = 0

for num in nd:
  if nd[num] > numeros[num]:
      print('não consigo')
      consn = 0
      break
if consd == 1 and consnd==1:
  print('consigo')
```

```python
# fazer a pulseira e diminuir quantidade de letras disponíveis

if consd == 1:
  print(palavrad)
  conf = input('fazer pulseira com essa palavra? digite s para sim e qualquer coisa para nao')
  if conf == 's':
    print('quantidades atualizadas: ')
    for letra in pd:
      dourada[letra]-+pd[letra]
      print(letra + ': ' + str(dourada[letra]))
  else:
    print('ok')
```
